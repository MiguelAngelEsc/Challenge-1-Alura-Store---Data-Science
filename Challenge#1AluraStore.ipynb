### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

df_completo = pd.concat([
    tienda.assign(Tienda="Tienda 1"),
    tienda2.assign(Tienda="Tienda 2"),
    tienda3.assign(Tienda="Tienda 3"),
    tienda4.assign(Tienda="Tienda 4")
])

df_completo
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

```python
ingresos = {
    "Tienda 1": ingreso_tienda1,
    "Tienda 2": ingreso_tienda2,
    "Tienda 3": ingreso_tienda3,
    "Tienda 4": ingreso_tienda4
}

tiendaMayorIngresos = max(ingresos, key=ingresos.get)

print(f"La tienda con mayores ingresos es: {tiendaMayorIngresos}")
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python

# Primero, agrupamos por Tienda y Producto para contar cuántas unidades se vendieron de cada uno
ventas = df_completo.groupby(["Tienda", "Producto"]).size().reset_index(name="Unidades")

#  Para los MÁS VENDIDOS (TOP):
# Calculamos cuál es el número máximo de ventas en cada tienda
max_por_tienda = ventas.groupby("Tienda")["Unidades"].transform("max")
# Filtramos las filas que tengan exactamente ese número máximo
top = ventas[ventas["Unidades"] == max_por_tienda]

#  Para los MENOS VENDIDOS (BOTTOM):
# Calculamos cuál es el número mínimo de ventas en cada tienda
min_por_tienda = ventas.groupby("Tienda")["Unidades"].transform("min")
# Filtramos las filas que tengan exactamente ese número mínimo
bottom = ventas[ventas["Unidades"] == min_por_tienda]

# Mostramos los resultados
print("--- Productos Más Vendidos ---")
display(top)
print("\n--- Productos Menos Vendidos ---")
display(bottom)
```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
# Costo de envío promedio por tienda
df_costo_envio = df_completo.groupby("Tienda")["Costo de envío"].mean().reset_index()
print(df_costo_envio)

# Gráfico de área con valores
plt.figure(figsize=(8,4))

# Área
plt.fill_between(df_costo_envio["Tienda"], df_costo_envio["Costo de envío"], alpha=0.4)

# Línea
plt.plot(df_costo_envio["Tienda"], df_costo_envio["Costo de envío"], marker="o")

# Títulos
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de Envío Promedio")
plt.grid(True)

# Mostrar valores en cada punto
for i, valor in enumerate(df_costo_envio["Costo de envío"]):
    plt.text(i, valor, f"{valor:.2f}", ha='center', va='bottom')

plt.show()
```

## 1. Introducción

El presente informe tiene como objetivo evaluar el desempeño comercial, operativo y de servicio de las cuatro tiendas de Alura Store Latam (Tiendas 1, 2, 3 y 4). La finalidad es identificar la unidad de negocio con menor rendimiento integral para recomendar su venta, permitiendo así liberar capital para reinvertir en las sucursales con mayor potencial de crecimiento.

La decisión se fundamenta en un análisis cuantitativo de cuatro pilares clave:
* **Facturación Total:** Capacidad de generación de ingresos.
* **Satisfacción del Cliente:** Calidad percibida a través de las calificaciones.
* **Eficiencia Logística:** Promedio de costos de envío.
* **Dinámica de Productos:** Identificación de productos más demandados y de bajo rendimiento.

---

## 2. Análisis de Resultados

### A. Ingresos y Facturación (El Factor Crítico)
Al analizar el flujo de caja, se observa una disparidad significativa entre las sucursales.

* **Líder en Ventas:** La **Tienda 1** se consolida como el motor económico del grupo, generando ingresos totales de **$1,150,880,400**.

* **Menor Desempeño:** La **Tienda 4** registra la facturación más baja, con un total de **$1,038,375,700**.

* **Brecha:** Existe una diferencia de más de **$112 millones** entre la tienda líder y la Tienda 4, lo que evidencia una menor captación de mercado por parte de esta última.

### B. Satisfacción del Cliente (Calidad de Servicio)
La reputación de la marca varía según la ubicación:

* **Mejor Experiencia:** La **Tienda 3** lidera en satisfacción con una calificación promedio de aproximadamente **4.05**, lo que sugiere un equipo de ventas y atención al cliente altamente efectivo.
* **Áreas de Mejora:** La **Tienda 1**, a pesar de su alto volumen de ventas, tiene la calificación más baja (**3.98**), lo que indica que el volumen podría estar saturando su capacidad de servicio.
* **Situación de la Tienda 4:** Con un promedio de **3.99**, la Tienda 4 no logra destacar en servicio al cliente, manteniéndose en una zona de bajo impacto en cuanto a satisfacción se refiere.

### C. Eficiencia Logística
* **Costos Operativos:** La **Tienda 4** presenta el costo de envío promedio más bajo (**$23,459**). Si bien esto podría parecer una ventaja, al correlacionarlo con sus bajos ingresos, sugiere que la tienda está moviendo mercancía de menor valor o volumen, en lugar de ser más eficiente.

* **Comparativa:** La Tienda 1 asume los costos de envío más altos ($26,018), lo cual es coherente con su mayor actividad comercial y venta de productos de gran volumen (como muebles).

### D. Dinámica de Productos
El análisis de productos revela diferencias en la demanda local:

* **Portafolio Tienda 1:** Muestra una demanda robusta y diversificada con un triple empate en productos más vendidos (*Armario, Microondas, TV LED* con 60 unidades cada uno).
* **Portafolio Tienda 4:** Depende fuertemente de un solo producto (*Cama box* con 62 unidades), mientras que otros artículos como la *Guitarra eléctrica* tienen una salida mínima (33 unidades), indicando un inventario con menor rotación.

---

## 3. Conclusión y Recomendación

Tras analizar los factores financieros, operativos y de servicio, se recomienda proceder con la **VENTA DE LA TIENDA 4**.

### Justificación de la Decisión:

1.  **Bajo Rendimiento Financiero:** Es la unidad que menos aporta al flujo de caja total del grupo, con una diferencia sustancial respecto a las demás.
2.  **Falta de Ventaja Competitiva:** La Tienda 4 no destaca en ningún indicador clave. No es la que más vende (Tienda 1), ni la que mejor atiende (Tienda 3). Se encuentra en una posición de estancamiento.
3.  **Costo de Oportunidad:** Mantener una operación que no lidera en ingresos ni en satisfacción impide destinar recursos a potenciar las fortalezas de las otras sucursales.

### Estrategia Sugerida:
El capital obtenido de la venta debería reinvertirse con dos focos principales:
* **Optimización de la Tienda 1:** Mejorar su logística y capacitación de personal para elevar su calificación de servicio (actualmente 3.98) sin sacrificar su volumen de ventas.
* **Fidelización en la Tienda 3:** Reforzar las estrategias que han llevado a su excelencia en servicio para retener a su base de clientes leales.
